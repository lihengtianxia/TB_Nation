```python
inputfile = './data/nation_TB1.csv'
nation = pd.read_csv(inputfile)[['area','incidence','death','incidence rate','death rate']]
nation
```

```python
nation.index=range(2005,2015)
```

```python
l = ['incidence','death','incidence rate','death rate']
for k in range(2015,2036):
    nation.loc[k] = None
```

```python
nation
```

```python
for i in l:
    for j in range(2015,2036):
        f = GM11(nation[i].loc[range(2005,2015)].as_matrix())
        nation[i].loc[j] = f[0](j-2005+1)
nation
```

```python
nation['incidence_rate'] = None
for k in range(2005,2036):
    f = GM11(nation['incidence rate'].loc[range(2005,2015)].as_matrix())
    nation['incidence_rate'].loc[k] = f[0](k-2005+1).round(2)
nation
```

### 新陈代谢GM(1,1)

```python
nation_gm = nation
nation_gm['incidence_rate'] = None

for m in range(2005,2015):
    f = GM11(nation_gm['incidence rate'].loc[range(2005,2015)].as_matrix())
    nation_gm['incidence_rate'].loc[m] = f[0](m-2005+1).round(2)
    
count = 0 
for k in range(2015,2036):
    f = GM11(nation_gm['incidence rate'].loc[range(2005+count,2015+count)].as_matrix())
    nation_gm['incidence_rate'].loc[k] = f[0](k-2005+1).round(2)
    count += 1    
nation_gm
    
```

```python
count = 0
for k in range(2005,2036):
    if k < 2015:
        f = GM11(nation_gm['incidence rate'].loc[range(2005,2015)].as_matrix())
    else:
        f = GM11(nation_gm['incidence rate'].loc[range(2005+count,2015+count)].as_matrix())
        count += 1 
    nation_gm['incidence_rate'].loc[k] = f[0](k-2005+1).round(2)
nation_gm
```
