```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
os.chdir('/Users/Evan/DataScience/TB_Nation/Datasets/')
files = os.listdir()

datasets = pd.DataFrame()
for i in range(1,len(files)):
    for j in os.listdir(files[i]):
        for k in range(1,33):
            data = pd.read_excel(files[i]+'/'+ j ,skiprows=1).iloc[k:k+1,0:5]
            data = data.rename(columns={'Unnamed: 0':'Area','发病数':'Incidence','死亡数':'Death','发病率':'Incidence_rate','死亡率':'Death_rate'})
            data['Year'],data['Month'],data['Day'],data['Province']= files[i],j[4:6],'15',k
            datasets = pd.concat([datasets,data])
```

```python
datasets.sort_values(by=['Area','Month'])
```

```python
datasets['Date'] = datasets['Year'] + datasets['Month'] + datasets['Day']
datasets.index = pd.to_datetime(datasets['Date'])
datasets = datasets.drop(['Day'],axis=1)
datasets.head()
```

```python
datasets.to_csv('/Users/Evan/Desktop/Seasonal Variation2.csv')
```
